```python
import pandas
import sqlalchemy

engine = sqlalchemy.create_engine('postgresql+psycopg2://postgres:postgres123@localhost/nycdata')
```

```python

query = '''
SELECT 
	tpep_pickup_datetime::DATE as pickup_day,
	total_amount
FROM 
	public.yellow_taxis
WHERE
    tpep_pickup_datetime::DATE BETWEEN '2023-12-01' AND '2023-12-31'
order by
    1 asc
'''


df = pandas.read_sql_query(query, con=engine)
df
```

```python
plot_data = df.groupby('pickup_day')['total_amount'].sum()

import matplotlib
chart = plot_data.plot(xlabel="Pickup Day", ylabel="Total Amount", kind='line', legend=True, figsize=(12, 8))
```
