```python
import pandas
import sqlalchemy

engine = sqlalchemy.create_engine('postgresql+psycopg2://postgres:postgres123@localhost/nycdata')
```

```python

query = '''
SELECT 
	tpep_pickup_datetime::DATE as pickup_day,
	total_amount
FROM 
	public.yellow_taxis
'''


df = pandas.read_sql_query(query, con=engine)
df
```

```python
groups = df.groupby('pickup_day')

import matplotlib
chart = groups.total_amount.plot(xlabel="Sample", ylabel="Date", kind='line', legend=True, marker='o', linestyle='dashed')
```
